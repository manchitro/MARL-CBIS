```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_f1(train_prec, train_rec, eval_prec, eval_rec, title):
	# Calculate the F1 score
	train_f1 = 2 * (train_prec * train_rec) / (train_prec + train_rec)
	eval_f1 = 2 * (eval_prec * eval_rec) / (eval_prec + eval_rec)

	print(max(train_f1))
	print(max(eval_f1))

	# Create subplots for precision, recall, and F1 score
	fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(15, 5))
	fig.suptitle(title)

	# Plot train and eval precision
	ax1.plot(train_prec, label='train_prec')
	ax1.plot(train_prec.idxmax(), train_prec.max(), 'ro', label='train_max='+str(train_prec.max()))
	ax1.plot(eval_prec, label='eval_prec')
	ax1.plot(eval_prec.idxmax(), eval_prec.max(), 'ro', label='eval_max='+str(eval_prec.max()))
	ax1.set_xlabel('Epoch')
	ax1.set_ylabel('Precision')
	ax1.set_ylim([0, 1])
	ax1.legend()

	# Plot train and eval recall
	ax2.plot(train_rec, label='train_rec')
	ax2.plot(train_rec.idxmax(), train_rec.max(), 'ro', label='train_max='+str(train_rec.max()))
	ax2.plot(eval_rec, label='eval_rec')
	ax2.plot(eval_rec.idxmax(), eval_rec.max(), 'ro', label='eval_max='+str(eval_rec.max()))
	ax2.set_xlabel('Epoch')
	ax2.set_ylabel('Recall')
	ax2.set_ylim([0, 1])
	ax2.legend()

	# Plot train and eval F1 score
	ax3.plot(train_f1, label='train_f1')
	ax3.plot(train_rec.idxmax(), train_rec.max(), 'ro', label='train_max='+str(train_rec.max()))
	ax3.plot(eval_f1, label='eval_f1')
	ax3.plot(eval_rec.idxmax(), eval_rec.max(), 'ro', label='eval_max='+str(eval_rec.max()))
	ax3.set_xlabel('Epoch')
	ax3.set_ylabel('F1 Score')
	ax3.set_ylim([0, 1])
	ax3.legend()

	# Display the plot
	plt.show()

def plot_error(train_data, title):
	# Create figure with 3 subplots
	fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,5))
	fig.suptitle(title)

	# Set y-axis limit to go from 0.0 to 1.0 for all three graphs
	axs[0].set_ylim(bottom=0, top=20)
	axs[1].set_ylim(bottom=0, top=20)
	axs[2].set_ylim([0.0, 1.0])

	# Plot actor loss
	axs[0].plot(train_data['a_loss'])
	axs[0].set_title('Actor Loss')
	axs[0].set_xlabel('Epoch')

	# Plot critic loss
	axs[1].plot(train_data['c_loss'])
	axs[1].set_title('Critic Loss')
	axs[1].set_xlabel('Epoch')

	# Plot error
	axs[2].plot(train_data['error'])
	axs[2].set_title('Error')
	axs[2].set_xlabel('Epoch')

	# Show the plot
	plt.show()
```
